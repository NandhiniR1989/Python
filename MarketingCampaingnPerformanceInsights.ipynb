```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##**1. Load the Dataset**

```python
#Read the marketing campaign data from the CSV file into a pandas DataFrame.
df = pd.read_csv('https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv')
```

##**2. Descriptive Analysis**

###**Basic Structure**

```python
#Print the first few rows of the dataset to get an overview of the data.
df.head(5)
```

```python
#Obtain the number of rows and columns in the dataset.
df.shape
```

```python
#Get a concise summary of the dataset, including the data types and non-null values.
df.info()
```

```python
#Generate descriptive statistics for numerical columns.
df.describe().T
```

###**Data Exploration:**

```python
#Print the number of unique Campaign_ID values in the dataset.
df['Campaign_ID'].nunique()
```

```python
#List the unique values of the Location and Customer_Segment columns.
location = df['Location'].unique()
cust_seg= df['Customer_Segment'].unique()
print(f'Unique Location: \n{location}\n \nUnique Customer Segment: \n{cust_seg}')
```

```python
#Count the occurrences of each category in the Campaign_Type and Channel_Used and columns.
Type_Of_Campain = df.Campaign_Type.value_counts()
Used_Channel = df.Channel_Used.value_counts()
print(f'Count of Campaign_Type:\n\n{Type_Of_Campain}\n\nCount of Channel_Used: \n\n{Used_Channel}')
```

##**3. Exploratory Data Analysis (EDA) and Visualization**

###**Campaign Performance**

```python
#Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.
plt.figure(figsize=(10,5))
plt.scatter(x= 'Acquisition_Cost', y='ROI', data=df, color = 'Cyan', edgecolors= 'Black')
plt.title('Relationship B/W Acquisition_Cost and ROI', fontsize = 15, fontweight = 'bold')
plt.xlabel('Acquisition_Cost',fontsize=10,fontweight = 'bold')
plt.ylabel('ROI',fontsize=10, fontweight = 'bold')
plt.show()
```

```python
#Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.
plt.figure(figsize=(10,7))
sns.barplot(data=df, x = 'Channel_Used', y= 'Conversion_Rate', hue = 'Campaign_Type',estimator='mean',palette='rainbow')
plt.title('Average Conversion Rate for Channel Used and Campaign Type', fontweight = 'bold')
plt.xlabel('Channel Used', fontweight= 'bold')
plt.ylabel('Campaign Type', fontweight = 'bold')
plt.show()

```

```python
#Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.
plt.figure(figsize=(8,5))
sns.boxplot(x= 'Engagement_Score', y= 'Campaign_Type', hue = 'Campaign_Type', data=df, palette='pastel')
plt.title('Distribution of Engagement_Score across Different Campaign Type', fontsize = 12, fontweight = 'bold')
plt.xlabel('Engagement_Score', fontweight ='bold')
plt.ylabel('Campaign_Type', fontweight ='bold')
plt.show()
```

```python
#Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.
plt.figure(figsize=(8,4))
sns.barplot(x= df['Company'], y= df['ROI'], estimator = 'mean', color='lightgreen')
plt.title('Average ROI by Company', fontsize = 10, fontweight = 'bold')
plt.xticks(rotation =50)
plt.show()
```

```python
#Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.
correlation_matrix = df[['Engagement_Score', 'Conversion_Rate']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(correlation_matrix, annot=True, cmap='plasma')
plt.title('Corrrelation matrix', fontweight = 'bold')
plt.show()
```

###**Customer Segmentation:**

```python
#Create a count plot to visualize the distribution of Target_Audience.
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Target_Audience',color='orange')
plt.title('Target Audience Distribution', fontweight='bold')
plt.xlabel('Target_Audience', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.show()
```

```python
#Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.
plt.figure(figsize=(10,8))
sns.barplot(data=df, x='Customer_Segment', y='Conversion_Rate', hue='Language',estimator = 'max', palette= 'muted')
plt.title('Conversion Rate for each Language', fontsize=16, fontweight= 'bold')
plt.xlabel('Customer Segment', fontsize=12, fontweight='bold')
plt.ylabel('Conversion Rate', fontsize=12, fontweight='bold')
plt.show()
```

```python
#Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Acquisition_Cost', y='Customer_Segment', hue='Channel_Used',palette= 'Spectral')
plt.title('Distribution of Acquisition Cost across each Customer Segment by Channel Used', fontsize=10, fontweight= 'bold')
plt.xlabel('Acquisition Cost', fontsize=10, fontweight='bold')
plt.ylabel('Customer Segment',fontsize=10, fontweight='bold')
plt.show()
```

```python
#Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.
plt.figure(figsize=(10,8))
sns.barplot(data=df, x='Conversion_Rate', y='Language', hue='Campaign_Type', estimator='mean',palette= 'BuGn')
plt.title('Conversion Rate by Language to compare the effectiveness of campaigns', fontsize=16, fontweight= 'bold')
plt.xlabel('Conversion_Rate', fontsize=10, fontweight='bold')
plt.ylabel('Language', fontsize=10, fontweight='bold')
plt.show()
```

###**Channel Effectiveness:**

```python
#Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Channel_Used', y='Engagement_Score', hue = 'Campaign_Type', palette='YlOrRd')
plt.title('Compare Engagement Score with Channel Used and Segmented by Campaign Type', fontweight='bold')
plt.ylabel('Engagement Score', fontweight='bold')
plt.xlabel('Channel Used', fontweight='bold')
plt.show()
```

```python
#Show the distribution of total ROI across different Channel_Used using a pie chart.
ROI_Total = df.groupby('Channel_Used')['ROI'].sum()
plt.pie(ROI_Total,labels=ROI_Total.index, autopct='%1.2f%%')
plt.title('ROI Distribution across Channels_Used', fontweight = 'bold')
plt.show()

```

```python
#Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.
plt.figure(figsize=(12,6))
sns.scatterplot(data = df, x= 'Clicks',y= 'Impressions', hue= 'Channel_Used', palette='muted')
plt.title('Relationship between Clicks and Impressions', fontweight = 'bold')
plt.show()
```

###**Time-Based Analysis:**

```python
#Plot the distribution of Duration using a histogram.
plt.figure(figsize=(6,4))
sns.histplot(df['Duration'], kde=True)
plt.title('Duration Distribution', fontweight='bold')
plt.show()
```

```python
#Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Conversion_Rate',hue='Company',data=df, errorbar=None)
plt.title('Overall Conversion Rate Over Date for each Company',fontsize=14, fontweight = 'bold')
plt.xlabel('Date', fontsize=10, fontweight='bold')
plt.ylabel('Conversion_Rate', fontsize=10, fontweight='bold')
plt.show()
```

```python
#Examine the trend of Engagement_Score over Date with a line chart.
plt.figure(figsize=(10,4))
sns.lineplot(x='Date', y= 'Engagement_Score', data=df, errorbar=None)
plt.title('Trend of Engagement_Score Over Date',fontsize=14, fontweight = 'bold')
plt.xlabel('Date', fontsize=10, fontweight='bold')
plt.ylabel('Engagement_Score', fontsize=10, fontweight='bold')
plt.show()
```

###**Geographic Analysis:**

```python
#Determine which location has the highest Acquisition_Cost using a bar chart.
plt.figure(figsize=(10,6))
sns.barplot(x='Location', y='Acquisition_Cost', data =df ,estimator =sum, color='cyan')
plt.title('Acquisition_Cost as per Location',fontsize=14, fontweight = 'bold')
plt.xlabel('Location', fontsize=10, fontweight='bold')
plt.xticks(rotation=50)
plt.ylabel('Acquisition_Cost', fontsize=10, fontweight='bold')
plt.show()
```

```python
#Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.
plt.figure(figsize=(10,6))
sns.barplot(x='Location', y= 'Conversion_Rate', hue= 'Target_Audience',data=df, palette='viridis')
plt.title('Conversion_Rate by Location and Target_Audience',fontsize=14, fontweight = 'bold')
plt.xlabel('Location', fontsize=10, fontweight='bold')
plt.xticks(rotation=30)
plt.ylabel('Conversion_Rate', fontsize=10, fontweight='bold')
plt.show()
```

```python
#Illustrate the proportion of ROI by Location using a pie chart.
ROI = df.groupby('Location')['ROI'].sum()
plt.pie(ROI,labels=ROI.index, autopct='%1.2f%%')
plt.title('Proportion of ROI by Location', fontweight = 'bold')
plt.show()
```
